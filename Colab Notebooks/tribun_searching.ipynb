```
import requests
def getUrl(url):
    i=0

    '''ambil url tiap berita di indeks'''

    tautan = []

    page = requests.get(url)
  
    # parse html content
    soup = BeautifulSoup( page.content , 'html.parser')
      
    # get all tags
    tags = {tag.name for tag in soup.find_all()}
      
    # iterate all tags
    for tag in tags:
      
        # find all element of tag
        for i in soup.find_all( tag ):
      
            # if tag has attribute of class
            if i.has_attr( "class" ):
              print(i['class'])
              if i['class'] == "gs-title":
                
                for link in i.find('a'):
                  isi = link.get('href')
                  print(isi)
                  tautan.append(isi)
                  print(tautan)
                  print (len(tautans))
                return tautan
```
