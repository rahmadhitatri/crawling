```
def getUrl(url):

    '''ambil url tiap berita di indeks'''

    tautan = []

    url = urllib.request.urlopen(url)



    result = url.read()

    url.close()

    soup = BeautifulSoup(result, "html.parser")



    for link in soup.find_all('h3'):

        for l in link.find_all('a'):

            isi = l.get('href')

            tautan.append(isi)

    return tautan
```

```
def getIndeks(url):

    i = 1

    tautans = []

    tautan = getUrl(url)

    while (len(tautan) != 0 and i<=20):
        url = "https://m.tribunnews.com/tag/hoax?page=%d" %i
        #url = "https://m.tribunnews.com/index-news?date=%s" %day + "&page=%d" %i

        print (url)

        i += 1

        tautan = getUrl(url)

        tautans += tautan



    print (len(tautans))

    bf = open(filelink, "w")

    for t in tautans:

        bf.write(t)

        bf.write("\n")

    bf.close()
```
